```python
import pandas as pd
import json
```

```python
ob = pd.read_excel('./scratch/obce_obyv.xlsx')
```

```python
ob = ob[ob.obyv == ob.obyv]
```

```python
ob.kodob = ob.kodob.astype('int')
```

```python
vel = {
    'vesnice': (0, 1999),
    'malomesto': (2000, 9999),
    'stredomesto': (10000, 49999),
    'velkomesto' : (50000, 2000000) #244088
}
```

```python
def cat(obyv):
    if obyv <= 1999:
        return 'vesnice'
    elif 2000 <= obyv <= 9999:
        return 'malomesto'
    elif 10000 <= obyv <= 49999:
        return 'stredomesto'
    elif 50000 <= obyv:
        return 'velkomesto'
    else:
        print('varn')
```

```python
ob['kat'] = ob.obyv.apply(lambda x: cat(x))
```

```python
out = {}
for v in vel:
    out[v] = list(ob[ob.kat == v].kodob.values)
```
